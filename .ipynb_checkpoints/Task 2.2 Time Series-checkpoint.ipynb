```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the given URL
data_url = "https://github.com/LuisM78/Appliances-energy-prediction-data/raw/master/energydata_complete.csv"
df = pd.read_csv(data_url)

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Display the first few rows of the dataset
print(df.head())
```

```python
# Task 1: Plot the appliances energy consumption for the whole period and a closer look at any one week of consumption
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(df['date'], df['Appliances'])
plt.title('Appliances Energy Consumption - Whole Period')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')

# Select data for exactly one week
sample_week_data = df.set_index('date')['2016-01-11':'2016-01-17']
plt.subplot(1, 2, 2)
plt.plot(sample_week_data.index, sample_week_data['Appliances'])
plt.title('Appliances Energy Consumption - Sample Week')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')

plt.tight_layout()
plt.show()
```

```python
# Task 2: Plot heatmap of hourly consumption of appliances for a week

# Create a new column for the day of the week
df['day_of_week'] = df['date'].dt.day_name()

# Plotting the heatmap
plt.figure(figsize=(15, 7))
hourly_data_days = df.groupby(['day_of_week', df['date'].dt.hour])['Appliances'].mean().unstack()

# Reorder days for proper ordering in the heatmap
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hourly_data_days = hourly_data_days.reindex(index=days_order)

sns.heatmap(hourly_data_days, cmap='coolwarm', linewidths=.5, annot=True, fmt=".1f", cbar_kws={'label': 'Energy Consumption'})
plt.title('Hourly Appliances Energy Consumption by Day of the Week')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()

```

```python
# Task 3: Plot the histogram of energy consumption of appliances
plt.figure(figsize=(12, 6))
plt.hist(df['Appliances'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of Appliances Energy Consumption')
plt.xlabel('Energy Consumption')
plt.ylabel('Frequency')
plt.show()
```

```python
# Task 4: Construct a feature variable NSM (no. of seconds from midnight) and plot energy consumption vs. NSM
# Create NSM feature
df['NSM'] = (df['date'] - df['date'].dt.normalize()).dt.total_seconds()

# Plotting energy consumption vs. NSM
plt.figure(figsize=(15, 5))
plt.scatter(df['NSM'], df['Appliances'], alpha=0.5, color='orange')
plt.title('Energy Consumption vs. No. of Seconds from Midnight (NSM)')
plt.xlabel('NSM (seconds)')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
# Task 5: Plot appliances energy consumption vs. Press mm Hg
# Plotting energy consumption vs. Press mm Hg
plt.figure(figsize=(15, 5))
plt.scatter(df['Press_mm_hg'], df['Appliances'], alpha=0.5, color='green')
plt.title('Energy Consumption vs. Pressure (mm Hg)')
plt.xlabel('Pressure (mm Hg)')
plt.ylabel('Energy Consumption')
plt.show()
```

## Task: 6: It is observed that the major contributing factors for the energy consumption among all other features is NSM and Press mm Hg. Comment on it.

### Observations on Major Contributing Factors for Energy Consumption

#### NSM (No. of Seconds from Midnight):
- This could be attributed to daily routines or specific times when appliances are more frequently used.
- For example, energy consumption might increase during morning and evening hours when people are likely to be at home, using appliances like heaters, cooking equipment, or lighting.

#### Press mm Hg (Atmospheric Pressure):
- Changes in atmospheric pressure could be associated with weather conditions.
- For instance, certain appliances like heating or cooling systems might be influenced by weather variations.
- An increase in energy consumption might be observed during periods of extreme weather conditions.
